```python
import numpy as np
import matplotlib.pyplot as plt
from pykrige.ok import OrdinaryKriging

# Step 1: Generate synthetic spatial data
np.random.seed(42)
x = np.random.uniform(0, 10, 15)  # Random x-coordinates
y = np.random.uniform(0, 10, 15)  # Random y-coordinates
z = np.sin(0.5 * x) + np.cos(0.5 * y) + np.random.normal(0, 0.1, len(x))  # Values at coordinates

# Step 2: Create an extended grid for prediction
grid_x = np.linspace(-2, 12, 100)  # Extended beyond data points
grid_y = np.linspace(-2, 12, 100)

# Step 3: Perform ordinary kriging
ok = OrdinaryKriging(
    x, y, z,
    variogram_model='spherical',  # Fits a spherical variogram model
    verbose=True,
    enable_plotting=False
)
z_pred, ss_pred = ok.execute('grid', grid_x, grid_y)

# Step 4: Extract variogram information
lags = ok.lags  # Distance bins
semivariance = ok.semivariance  # Computed semivariance values

# Step 5: Compute theoretical variogram values
# Extract model parameters
model_parameters = ok.variogram_model_parameters
sill = model_parameters[0]
range_ = model_parameters[1]
nugget = model_parameters[2]

# Define the spherical variogram model
def spherical_variogram(h, sill, range_, nugget):
    h = np.array(h)
    gamma = np.zeros_like(h)
    mask = h <= range_
    gamma[mask] = nugget + sill * (1.5 * (h[mask] / range_) - 0.5 * (h[mask] / range_) ** 3)
    gamma[~mask] = nugget + sill
    return gamma

theoretical_variogram = spherical_variogram(lags, sill, range_, nugget)

# Step 6: Plot the empirical semi-variogram and theoretical model
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.scatter(lags, semivariance, color='blue', label='Empirical')
plt.plot(lags, theoretical_variogram, color='red', label='Fitted Model')
plt.xlabel('Lag Distance')
plt.ylabel('Semivariance')
plt.title('Empirical Semi-Variogram')
plt.legend()

# Step 7: Plot the predicted surface
plt.subplot(1, 3, 2)
plt.contourf(grid_x, grid_y, z_pred, levels=20, cmap='viridis')
plt.colorbar(label='Predicted Value')
plt.title('Predicted Surface')
plt.xlabel('X')
plt.ylabel('Y')

# Step 8: Plot the prediction standard error
plt.subplot(1, 3, 3)
plt.contourf(grid_x, grid_y, np.sqrt(ss_pred), levels=20, cmap='viridis')
plt.colorbar(label='Standard Error')
plt.title('Prediction Standard Error')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```
